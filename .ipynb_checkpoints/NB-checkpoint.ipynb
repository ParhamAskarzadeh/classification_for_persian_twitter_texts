```python
use_cols=['Char', 'Neg', 'Pos']
emoji_df=pd.read_csv("ijstable.csv",usecols=use_cols)
emoji_df.dropna(inplace=True)
emoji_df['state']= emoji_df[['Neg','Pos']].apply(lambda x: '||neg||' if x['Neg']>x['Pos'] else '||pos||', axis=1)
emoji_df
```

```python
char = emoji_df['Char'].tolist()
state_char = emoji_df['state'].tolist()
emoji_dict = {}
for i,c in enumerate(char):
    emoji_dict[c] = state_char[i]
emoji_dict
```

```python
text = 'Ø³Ù„Ø§Ù… ğŸ˜‚ Ú†Ø·ÙˆØ±ÛŒ ğŸ˜©'
def emotions(text):
    for emoji in emoji_dict:
        text = re.sub(emoji, emoji_dict[emoji], text)
    return text
text = emotions(text)
text
```

```python
class NormalizerOfTwitter:

    def __url(self, content):
        # url --> ||url||
        return content['text'].apply(lambda x: re.sub(
            r"(?i)\b((?:https?://|www\d{0,3}[.]|[a-z0-9.\-]+[.][a-z]{2,4}/)(?:[^\s()<>]+|\(([^\s()<>]+|(\([^\s()<>]+\)))*\))+(?:\(([^\s()<>]+|(\([^\s()<>]+\)))*\)|[^\s`!()\[\]{};:'\".,<>?Â«Â»â€œâ€â€˜â€™]))",
            ' ||url|| ', str(x)))


    def __target(self, content):
        return content['text'].apply(lambda x: re.sub('(@.*) ', ' ||target|| ', str(x)))

    
    def __sequence_of_repeated_characters(self, content):
        def a(text):
            w = re.search(r'(\w)(?:\1{2,})', text)
            if w is None:
                return text
            return re.sub(w.group(), w.group()[0] * 2, text)

        content['text'].apply(lambda x: a(x))
        return content['text']


    def __tag_of_html(self, content):
        tags_of_html = ['<!â€“â€¦â€“>', '<!DOCTYPE>', '<a>', '<abbr>', '<address>', '<area>', '<article>', '<aside>',
                        '<audio>', '<b>', '<base>', '<bdi>', '<bdo>', '<blockquote>', '<body>', '<br>', '<button>',
                        '<canvas>', '<caption>', '<cite>', '<code>', '<col>', '<colgroup>', '<data>', '<datalist>',
                        '<dd>', '<del>', '<details>', '<dfn>', '<dialog>', '<div>', '<dl>', '<dt>', '<em>', '<embed>',
                        '<fieldset>', '<figcaption>', '<figure>', '<figure>', '<footer>',
                        '<form>', '<h1>', '<h2>', '<h3>', '<h4>', '<h5>', '<h6>', '<head>', '<header>', '<hr>',
                        '<html>', '<i>', '<iframe>', '<img>',
                        '<input>', '<ins>', '<kbd>', '<label>', '<legend>', '<fieldset>',
                        '<li>', '<link>', '<main>', '<map>', '<mark>', '<menu>', '<meta>', '<meter>', '<nav>',
                        '<noscript>', '<object>', '<ol>', '<optgroup>', '<option>', '<output>', '<p>',
                        '<param>', '<object>', '<picture>', '<pre>', '<progress>', '<q>', '<rp>',
                        '<rt>', '<ruby>', '<s>', '<samp>', '<script>', '<section>', '<select>', '<small>', '<source>',
                        '<span>', '<strong>', '<style>', '<sub>', '<summary>', '<details>',
                        '<sup>', '<svg>', '<table>', '<tbody>', '<td>', '<template>', '<textarea>', '<tfoot>', '<th>',
                        '<thead>', '<time>', '<title>', '<tr>', '<track>', '<u>', '<ul>', '<var>',
                        '<video>', '<wbr>', '&nbsp']

        def a(text):
            for tag in tags_of_html:
                text = re.sub(tag, '', text)

        # return re.sub('<.*>', '', text)
        content['text'].apply(lambda x: a(x))
        return content['text']


    def __emotions_emoji(self, content):
        # emotions (pos, neg) --> ||pos||, ||neg||
        def emotions(text):
            try:
                text = re.sub(':\)', '||pos||', text)
                text = re.sub(':\(', '||neg||', text)
            except:
                pass
            return text

        content['text'].apply(lambda x: emotions(x))
        return content['text']

    def __acronyms(self, content):
        def a(text):
            return re.sub('Ù¾ Ù†', 'Ù¾ÛŒ Ù†ÙˆØ´Øª', text)
        content['text'].apply(lambda x: a(x))
        return content['text']

    def __negation_characters(self, content):
        # not complete
        content['text'].apply(lambda x: re.sub('((?:Ù†Ù…ÛŒ|Ù†Ø®ÙˆØ§Ø³Øª|Ù†ØªÙˆØ§Ù†|Ù†Ø±ÙØª|Ù†Ø®ÙˆØ±Ø¯|Ù†Ø´Ø¯)[^ \.?!!ØŸ\n\r]*)', ' ||not|| ', str(x)))
        return content['text']
    
    def __remove_additional_space(self, content):
        # not complete
        content['text'].apply(lambda x: re.sub(r' {2,}', ' ', str(x)))
        return content['text']

    def __emotions_emoji2(self, content):
        def emotions(text):
            for emoji in emoji_dict:
                text = re.sub(emoji, emoji_dict[emoji], text)
        content['text'].apply(lambda x: emotions(x))
        return content['text']
        
        
    def __remove_emoji(self, content):
        content['text'].apply(lambda x: re.sub(r'[^\x00-\x7F]+', '', str(x)))
        return content['text']
    
    
    # def __case(self, content):
    #     content['text'].apply(lambda x: x.lower())
    #     return content['text']

    # def __unicode(self, text: str):
    #     # not for persian
    #     return re.sub('', '', text)
    

    def apply_all_filter(self, content):
        content['text'] = self.__target(content)
        content['text'] = self.__url(content)
        content['text'] = self.__sequence_of_repeated_characters(content)
        content['text'] = self.__tag_of_html(content)
        content['text'] = self.__negation_characters(content)
        content['text'] = self.__acronyms(content)
        content['text'] = self.__emotions_emoji(content)
        content['text'] = self.__emotions_emoji2(content)
        content['text'] = self.__remove_additional_space(content)
#         content['text'] = self.__remove_emoji(content)

        # content['text'] = self.__case(content)
        # content['text'] = self.__unicode(content)
        return content
```

```python
data = NormalizerOfTwitter().apply_all_filter(data)
data
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(bow, data['vector_approach'], test_size=0.2, random_state=11)
y_train
```

```python
from sklearn.naive_bayes import GaussianNB, BernoulliNB, MultinomialNB
from sklearn import metrics, tree
from sklearn.neighbors import KNeighborsClassifier

nb_model = MultinomialNB(alpha=0.1)
# nb_model = tree.DecisionTreeClassifier()
```

```python
nb_model.fit(x_train, y_train)
```

```python
y_pred = nb_model.predict(x_test)
```

```python
from sklearn.metrics import classification_report

print('Accuracy:', metrics.accuracy_score(y_test, y_pred))

print(classification_report(y_test, y_pred))
```
